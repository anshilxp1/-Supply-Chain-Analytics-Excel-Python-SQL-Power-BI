# cleaning

```python
df_orders.columns
```

```python
df_orders["Sales"]=df_orders["Sales"].round(2)
```

```python
df_orders["Benefit per order"]=df_orders["Benefit per order"].round(2)
```

```python
df_orders["Order Item Discount"]=df_orders["Order Item Discount"].round(2)
```

```python
df_orders["Order Item Product Price"]=df_orders["Order Item Product Price"].round(2)
```

```python
df_orders["Order Item Total"]=df_orders["Order Item Total"].round(2)
```

```python
df_orders.drop(columns=['Latitude','Longitude','Order Item Profit Ratio','Order Profit Per Order'],inplace=True)
```

```python
df_orders=df_orders.rename(columns={"order date (DateOrders)":"order date"})
```

```python
df_orders=df_orders.rename(columns={"Order Item Product Price":"price per item"})
```

```python
df_orders=df_orders.rename(columns={"Sales":"Total Sales"})
```

```python
#df_orders['order date'] = pd.to_datetime(df_orders['order date'],errors='coerce')
```

```python
#df_orders['order date']=df_orders['order date'].dt.date
```

```python
df_orders
```

```python
# Convert mixed date formats
df_orders["order date"] = pd.to_datetime(
    df_orders["order date"],
    format="mixed", # take all dates in one format automaticaly
    errors="raise"
)

# Remove the time part
df_orders["order date"] = df_orders["order date"].dt.date

# Check null values
print(df_orders["order date"].isnull().sum())
```

```python
# Create MySQL connection 
engine = create_engine("mysql+pymysql://root:12345@127.0.0.1:3306/supply_data")

# Push all tables to MySQL
df_category.to_sql('category', con=engine, if_exists='replace', index=False)
df_customer.to_sql('customers', con=engine, if_exists='replace', index=False)
df_orders.to_sql('orders', con=engine, if_exists='replace', index=False)
df_product.to_sql('products', con=engine, if_exists='replace', index=False)
df_shipping.to_sql('shipping', con=engine, if_exists='replace', index=False)

print("All tables loaded successfully!")
```
